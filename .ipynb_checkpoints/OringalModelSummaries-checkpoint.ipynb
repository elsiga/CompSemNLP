```python
import torch
from transformers import BartForConditionalGeneration, BartTokenizer
from datasets import load_dataset
import json
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
dataset = load_dataset('openai/summarize_from_feedback', 'comparisons')
```

```python
def generate_summary(model, tokenizer, text):
    inputs = tokenizer.encode("Summarize the following text in 30 words or less:"+text, return_tensors='pt', max_length=1024, truncation=True).to(device)
    summary_ids = model.generate(inputs, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True, max_new_tokens=55)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
model_name = 'facebook/bart-base'
bart_model = BartForConditionalGeneration.from_pretrained(model_name).to(device)
bart_tokenizer = BartTokenizer.from_pretrained(model_name)
```

```python
# validation dataset from openai/summarize_from_feedback
val_data = [item for item in dataset['validation']]
val_text = [item['info']['post'] for item in val_data]
```

```python
# TESTS checking summaries from different small models
print(val_text[500])

bart_summary = generate_summary(bart_model, bart_tokenizer, val_text[500])
print("\n\n\n" + bart_summary)
```

```python
original_summaries = [generate_summary(bart_model, bart_tokenizer, text) for text in val_text]

file_name = "original_bart_summaries.json"

with open(file_name, 'w') as json_file:
    json.dump(original_summaries, json_file, indent=4)
```
