```python
import pandas as pd
from datasets import load_dataset

# Load the dataset
dataset = load_dataset("openai/summarize_from_feedback",'comparisons')

# Initialize lists to store new columns
prompts = []
chosen_answers = []
rejected_answers = []

# Process each entry in the dataset
dataset_train = dataset['train']
dataset_val = dataset['validation']
for entry in dataset_train:
    post = entry['info']['post']
    summary1 = entry['summaries'][0]['text']
    summary2 = entry['summaries'][1]['text']
    # Determine which summary was preferred
    if entry['choice'] == 0:
        chosen = summary1
        rejected = summary2
    else:
        chosen = summary2
        rejected = summary1
    
    # Construct the new columns
    prompts.append(f"Write a TLDR for the following text: {post}")
    chosen_answers.append(chosen)
    rejected_answers.append(rejected)

# Create a DataFrame with the new columns
reformatted_data = pd.DataFrame({
    "Prompt": prompts,
    "Chosen": chosen_answers,
    "Rejected": rejected_answers
})

# Save the reformatted dataset to a CSV file
reformatted_data.to_csv("reformatted_summarize_from_feedback_train.csv", index=False)

# Display the first few rows of the reformatted dataset
reformatted_data.head()
```
